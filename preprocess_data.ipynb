```python
import numpy as np
import csv

with open('CoST.csv','r') as f:
    data = f.read().splitlines()
    
len(data)
```

```python
data[1]
```

```python
data[2]
```

```python
channels = []
labels = []

gesture = []
for d in data[1:]:
    d = d.split(',')
    if int(d[3].strip()) == 1: #frame
        channels.append(gesture)
        label = '{}{}'.format(int(d[1]),int(d[2]))
        labels.append(label)
        gesture = []
    tmp = [int(e.strip()) for e in d[4:]]
    gesture.append(tmp)
label = '{}{}'.format(int(d[1]),int(d[2]))
channels.append(gesture)
labels.append(label)
channels = channels[1:]
labels = labels[1:]
```

```python
print(len(channels))
print(len(labels))
```

```python
import json
with open('CoSTdata.json','w') as f:
    json.dump(channels,f)
with open('labels.json','w') as f:
    json.dump(labels,f)
```

```python
for i,c in enumerate(channels):
    print(len(c),labels[i])
```

```python
train_chan = []
train_lab = []
test_chan = []
test_lab = []
for i,ch in enumerate(channels):
    if i%5 == 0:
        test_chan.append(ch)
        test_lab.append(labels[i])
    else:
        train_chan.append(ch)
        train_lab.append(labels[i])
```

```python
print(len(train_chan),len(train_lab))
print(len(test_chan),len(test_lab))
```

```python
with open('traindata_cost.json','w') as f:
    json.dump(train_chan,f)
with open('trainlabels_cost.json','w') as f:
    json.dump(train_lab,f)
with open('testdata_cost.json','w') as f:
    json.dump(test_chan,f)
with open('testlabels_cost.json','w') as f:
    json.dump(test_lab,f)
```
